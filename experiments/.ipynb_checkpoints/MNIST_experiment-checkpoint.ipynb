```python
%matplotlib notebook

import sys
sys.path.append('..')
sys.path.append('../model/')

import utils
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
import numpy


[x,y] = utils.load_mnist(path='../datasets/mnist/')
m = utils.load_single('MNIST_0_model.zip')

output = m.get_hidden(x)

plt.figure()
color=cm.rainbow(numpy.linspace(0,1,10))
for l,c in zip(range(10),color):
    ix = numpy.where(y==l)[0]
    plt.scatter(output[ix,0],output[ix,1],c=c,label=l,s=8,linewidth=0)
# plt.xlim([-5.0,5.0])
# plt.ylim([-5.0,5.0])
plt.legend(fontsize=15)
plt.show()
# plt.savefig('z_epoch' + str(epoch) + '.pdf')
```
